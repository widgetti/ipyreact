```python
import ipyreact
from traitlets import  Int, Dict, observe
```

```python
class UpdateDictFromPythonWidget(ipyreact.ReactWidget):
    #note that when we add these traitlets, they will automatically be made available to the react component
    simple_dict = Dict({'foo':5, 'bar':8, 'baz':10}).tag(sync=True)
    count = Int(0).tag(sync=True)

    @observe('count')
    def _observe_count(self, change):
        new_val = self.count
        sd = self.simple_dict.copy()
        sd['foo'] += 1
        self.simple_dict = sd

    _esm = """
    import confetti from "canvas-confetti";
    import * as React from "react";
    
    export const DictView = ({fullDict}) => {
        return (<table>
                  <tr><th>foo</th><th>bar</th><th>baz</th></tr>
                  <tr><td>{fullDict['foo']}</td><td>{fullDict['bar']}</td><td>{fullDict['baz']}</td></tr>
                </table>);
    }

    export default function({on_count, debug, count, simple_dict}) {
        return <div><button onClick={() => confetti() && on_count(count + 1)}>
            {count} times confetti
        </button>
        <DictView fullDict={simple_dict} />
    </div>
    };"""
w = UpdateDictFromPythonWidget()
w
```
